```python
# 필요한 라이브러리 import
import pdfplumber
import os
import pandas as pd

# pdf 파일 열기
path = "/content/drive/MyDrive/Colab Notebooks/workplace/ca-warn-report.pdf"
pdf = pdfplumber.open(path)
```

```python
#start_page, convert pdf to image file
p0 = pdf.pages[0]
im = p0.to_image()
im
```

```python
#Use .extract_table to get the data from the largest table on the page
#.extract_table returns a list of lists, with each inner list representing a row in the table. Here are the first three rows:
table = p0.extract_table()
table[:3]
```

```python
#basic cleanup
import pandas as pd
df = pd.DataFrame(table[1:], columns=table[0])
for column in ["Effective", "Received"]:
    df[column] = df[column].str.replace(" ", "")
```
